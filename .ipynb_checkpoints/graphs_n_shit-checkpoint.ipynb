```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

RANDOM_STATE=42
```

```python
#load datasets
data_playlist = pd.read_csv('yaren_data_analysis/data/playlist_tracks.csv')
recommended_playlist = pd.read_csv('yaren_data_analysis/data/recommended_tracks.csv')
```

```python
#displays the first 5 songs on your own playlist
data_playlist.head()
```

```python
#displays the first 5 songs of recommended songs
recommended_playlist.head()
```

```python
#Calculate the minimum, mean, and maximum for the playlist tracks and recommneded tracks for each column

#adding new temp columns
recommended_means = {}
user_means = {}

print("------ Recommended Playlist ------\n")
#recommended playlists
for column in recommended_playlist.columns[3:]:
    feature_mean = recommended_playlist[column].mean()
    recommended_means[column] = feature_mean #for bar graph
    print(f"Mean of {column} in Recommended Songs: {feature_mean}")
print("\n------ User Playlist ------\n")
#we would want to do the same thing for our playlist
for column in data_playlist.columns[3:]:
    feature_mean = data_playlist[column].mean()
    user_means[column] = feature_mean #for bar graph
    print(f"Mean of {column} in Recommended Songs: {feature_mean}")


#visual to compare the averages between the recommended playlist and user playlist
features = list(user_means.keys())

for feature in features:
    recommended_mean = recommended_means[feature]
    user_mean = user_means[feature]

    fig, ax = plt.subplots(figsize=(8, 6))

    x = np.array([1])
    ax.bar(x - .35/2, recommended_mean, .35, label='Recommended Playlist')
    ax.bar(x + .35/2, user_mean, .35, label='User Playlist')

    #label the graphs
    ax.set_xlabel('Feature')
    ax.set_ylabel('Mean Values')
    ax.set_title(f'Comparison of {feature} Averages')
    
    ax.set_xticks(x)
    ax.set_xticklabels([feature], rotation=45, ha="right")  #rotate feature readability on the graph
    
    ax.legend() #legend
    
    plt.show()

plt.tight_layout()    
```

```python
#Pseudo Matrix Inversion Linear Regression
# define matrix A and vector y
A = data_playlist[['Danceability']].values
y = data_playlist[['Tempo']].values

#psuedo-inverse of A
A_pseudo_inverse = np.linalg.pinv(A)

#compute the vector x
x = np.dot(A_pseudo_inverse, y)

# Print the computed x values rounded to 2 decimal places
computed_x = np.round(x, 2)
print("Computed x using pseudo-inverse of A: ", np.round(x, 2))

#Plot the data points, regression line, and set labels and title
plt.figure(figsize=(10, 6))
plt.scatter(A, y, color='#ff8200', label='Data points')
plt.plot(A, np.dot(A, x), color='red', label='Regression line')
plt.xlabel('Danceability')
plt.ylabel('Tempo')
plt.title('Linear Regression using Pseudo-Inverse')
plt.legend()
plt.show()
```

```python
#create a DataFrame with specified columns: danceability, tempo, energy, loudness
df = pd.DataFrame(data=data_playlist, columns=['Danceability', 'Tempo', 'Energy','Loudness'])
df.head()

#Features and target variables
X = np.array(df[['Tempo', 'Energy','Loudness']])
y = df['Danceability']

#split the data into training and testing sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=RANDOM_STATE)

#printing out the testing and training sets
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
#scaling the feature data to make sure that all features contribute qually to the model
scaler = StandardScaler()
#scale the training data
X_train_scaled = scaler.fit_transform(X_train)

#scale the test data
X_test_scaled = scaler.transform(X_test)
```

```python
#training a K-nearest Neighbors Regressor
n_neighbors_list = [1, 3, 5, 10, 15, 25, 35]
best_r2 = -float('inf')
best_k = None

#iterate over each value of n_neighbors
for k in n_neighbors_list:
    #create and train the KNN
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    
    #make predictions on the test set
    y_pred = knn.predict(X_test_scaled)
    
    #calculate r^2 score
    r2 = r2_score(y_test, y_pred)
    
    #print out the r^2 scores for the current value of n_neighbors
    print(f"n_neighbors: {k}, R^2 Score: {r2:.2f}")
```

```python
#the best r^2 score from the code above is 15
#it has the highest score
# Set the best k value found from the previous test
best_k = 15

#initialize the K-Nearest Neighbors Regressor with the best k value
best_model = KNeighborsRegressor(n_neighbors=best_k)

#fit the model using the scaled training data
best_model.fit(X_train_scaled, y_train)

#make predictions on the scaled test data using the trained KNN model
y_pred = best_model.predict(X_test_scaled)

#calculate the Mean Squared Error (MSE) between the predicted and actual values
mse_knn = mean_squared_error(y_test, y_pred)

#calculate the R^2 score for the model
r2_knn = r2_score(y_test, y_pred)

#print the MSE and R^2 score rounded to 2 decimal places
print(f"Mean Squared Error: {mse_knn:,.2f}")
print(f"R^2 Score: {r2_knn:.2f}")
```
